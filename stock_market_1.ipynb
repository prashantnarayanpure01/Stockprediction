```python
import pandas as pd                                      #importing pandas library
import numpy as np                                       #importing numpy library
from sklearn import linear_model                         #importing linear_model from scikit learn
from sklearn.model_selection import train_test_split     #importing train_test_split from scikit learn 
import matplotlib.pyplot as plt     
```

```python
df = pd.read_csv('stock_market_1.csv')                     #defining dataframe by extracting data from stock_market.csv
```

```python
df
```

```python
plt.plot(df['Open'],df['Close'])
```

```python
plt.scatter(df['Open'],df['Close'],marker='+',color='red')
```

```python
x = df[['Open']]
y = df[['Close']]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)                     #training the model
```

```python
model.coef_                                    #asking the model about coefficients
```

```python
model.intercept_                               #asking the model about intercept
```

```python
y_predicted = model.predict(x_test)                          #prediction of testing data
```

```python
y_predicted
```

```python
model.score(x_test,y_test)    #calculating similarity percentage between testing output from model and the original testing data
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_predicted)                  #RMSE value
```

# 99.48% Accuracy
